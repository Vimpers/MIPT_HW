```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Загрузка данных
images = np.load("C:/Users/Petroo/Desktop/DL2/images.npy")
labels = np.load("C:/Users/Petroo/Desktop/DL2/labels.npy")
images_sub = np.load("C:/Users/Petroo/Desktop/DL2/images_sub.npy")

# Нормализация изображений
images = images / 255.0
images_sub = images_sub / 255.0

# Преобразование меток
labels = to_categorical(labels, num_classes=26)

# Разделение данных на тренировочные и проверочные
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

# Создание модели
model = Sequential([
    Conv2D(64, (3, 3), activation='relu', input_shape=(48, 48, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(256, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(512, (3, 3), activation='relu'),  # Добавлен новый слой
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(26, activation='softmax')
])

# Компиляция модели
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Обучение модели без использования аугментации данных
batch_size = 128
model.fit(
    X_train, y_train,
    batch_size=batch_size,
    epochs=12,  # Установлено 12 эпох
    validation_data=(X_val, y_val)
)

# Предсказания на тестовом наборе данных
predictions = model.predict(images_sub)
predicted_labels = np.argmax(predictions, axis=1)


```

```python
# Создание DataFrame для сабмита
submission = pd.DataFrame({
    'Id': np.arange(len(predicted_labels)),
    'Category': predicted_labels
})

# Сохранение результатов в CSV файл в текущей директории
submission.to_csv('submission.csv', index=False)

# Сохранение результатов в CSV файл в указанной директории
submission.to_csv("C:/Users/Petroo/Desktop/DL2/sample_submission.csv", index=False)
```
